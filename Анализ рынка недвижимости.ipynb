```python
def change_letter(row):
    need_change = row['locality_name']
    table = str.maketrans('ё', 'е')
    return need_change.translate(table)
```

```python
clean_estate_data['locality_name'] = clean_estate_data.apply(change_letter, axis = 1)
len(clean_estate_data['locality_name'].unique())
```

Выделим несколько отдельных категорий:

- группа 1 - Санкт-Петербург;
- группа 2 - города районного значения, для которых не указаны статусы "деревня" или "поселок" и т.д.
- группа 3 - маленькие населенные пункты: "деревня", "поселок", "товарищество", "село". 

```python
def group_locality(row):

    locality = row['locality_name']
    if 'Санкт-Петербург' in locality:
        return 'Санкт-Петербург'
    
    if 'деревня' in locality:
        return 'малый населенный пункт'
    
    if 'поселок' in locality:
        return 'малый населенный пункт'
    
    if 'товарищество' in locality:
        return 'малый населенный пункт'
    
    if 'село' in locality:
        return 'малый населенный пункт'
    
    else:
        return 'город районного значения'
```

```python
clean_estate_data['locality_name_group'] = clean_estate_data.apply(group_locality, axis = 1)
```

Если хочется сэкономить на недвижимости, то студия или открытая планировка - то, что доктор прописал: недалеко от центра, маленькие, но относительно недорогие.
