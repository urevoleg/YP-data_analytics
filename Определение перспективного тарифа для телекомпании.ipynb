### Таблица *users* (информация о пользователях):

- **user_id** — уникальный идентификатор пользователя
- **first_name** — имя пользователя
- **last_name** — фамилия пользователя
- **age** — возраст пользователя (годы)
- **reg_date** — дата подключения тарифа (день, месяц, год)
- **churn_date** — дата прекращения пользования тарифом (если значение пропущено, то тариф ещё действовал на момент выгрузки данных)
- **city** — город проживания пользователя
- **tarif** — название тарифного плана

### Таблица *calls* (информация о звонках):

- **id** — уникальный номер звонка
- **call_date** — дата звонка
- **duration** — длительность звонка в минутах
- **user_id** — идентификатор пользователя, сделавшего звонок

### Таблица *messages* (информация о сообщениях):

- **id** — уникальный номер сообщения
- **message_date** — дата сообщения
- **user_id** — идентификатор пользователя, отправившего сообщение

### Таблица *internet* (информация об интернет-сессиях):

- **id** — уникальный номер сессии
- **mb_used** — объём потраченного за сессию интернет-трафика (в мегабайтах)
- **session_date** — дата интернет-сессии
- **user_id** — идентификатор пользователя

### Таблица *tariffs* (информация о тарифах):

- **tariff_name** — название тарифа
- **rub_monthly_fee** — ежемесячная абонентская плата в рублях
- **minutes_included** — количество минут разговора в месяц, включённых в абонентскую плату
- **messages_included** — количество сообщений в месяц, включённых в абонентскую плату
- **mb_per_month_included** — объём интернет-трафика, включённого в абонентскую плату (в мегабайтах)
- **rub_per_minute** — стоимость минуты разговора сверх тарифного пакета (например, если в тарифе 100 минут разговора в месяц, то со 101 минуты будет взиматься плата)

- **rub_per_gb** — стоимость дополнительного гигабайта интернет-трафика сверх тарифного пакета (1 гигабайт = 1024 мегабайта)

<span style="color: #9ACD32">***Users***</span><br>

- churn_date - выделим группу, где котракт был расторгнут, а после изменим тип данных
- reg_date - изменим тип данных

```python
users['churn_date'] = users['churn_date'].fillna(value = '2025-12-31')
users.head()
```

```python
def churn_event(row):
    """"определяет, совершилось ли расторжение контракта"""
    churn = row['churn_date']
    if churn == '2025-12-31':
        return 'действующий'
    
    return 'расторгнут'
```

```python
users['churn_event'] = users.apply(churn_event, axis = 1)
users['churn_date'] = pd.to_datetime(users['churn_date'],  format='%Y-%m-%d')
users['reg_date'] = pd.to_datetime(users['reg_date'], format = '%Y-%m-%d')
users.info()
```

## <span style="color: #228B22">3. Дополнение таблиц</span><br>

Для дальнейшей эффективной работы, нам необходимо свести всю имеющуюся информацию в один датасет, где по каждому отдельному пользователю, в зависимости от месяца, можно будет посмотреть статистику. В первую очередь, сгруппируем данные по совпадащим значениям -id и месяцу оказания той или иной услуги.

```python
calls_pivot = calls.pivot_table(index = ('user_id', 'month'), values = 'duration', aggfunc =  'sum').reset_index()
calls_pivot.columns = ['user_id', 'month',  'calls_sum']
print(len(calls_pivot['user_id'].unique()))
calls_pivot.head()
```

```python
messages_pivot = messages.pivot_table(index = ('user_id', 'month'), values = 'id', aggfunc = 'count').reset_index()
messages_pivot.columns = ['user_id', 'month', 'messages_count']
print(len(messages_pivot['user_id'].unique()))
print(len(messages_pivot))
messages_pivot.head()
```

```python
internet_pivot = internet.pivot_table(index = ('user_id', 'month'), values = 'mb_used', aggfunc =  'sum').reset_index()
internet_pivot.columns = ['user_id', 'month',  'mb_sum']
print(len(internet_pivot['user_id'].unique()))
internet_pivot.head()
```

```python
plump_data = users.merge(internet_pivot, on = 'user_id', how = 'outer')
print(len(plump_data['user_id'].unique()))
print(len(plump_data))
```

```python
plump_data_2 = plump_data.merge(calls_pivot, on = ('user_id', 'month'), how = 'outer')
print(len(plump_data_2['user_id'].unique()))
print(len(plump_data_2))
```

```python
complete_data = plump_data_2.merge(messages_pivot, on = ('user_id', 'month'), how = 'outer')
print(len(complete_data['user_id'].unique()))
print(len(complete_data))
```

Теперь рассчитаем выручку с каждого отдельно взятого клиента. 

```python
def overpayment_mb (row):
    
    """рассчитывает доход, сверх тарифного плана (интернет)"""
    
    mb_overrun = row['mb_overrun']
    rub_per_gb = row['rub_per_gb']
    
    if mb_overrun > 0:
       
        return (mb_overrun / 1024.0) * rub_per_gb
    
            
    else:
        
        return 0
         
    
def overpayment_calls (row):
    
    """рассчитывает доход, сверх тарифного плана (звонки)"""
    
    calls_overrun = row['calls_overrun']
    rub_per_minute = row['rub_per_minute']
    
    if calls_overrun > 0:
              
        return calls_overrun * rub_per_minute
    
    
    else:
        
        return 0    
    
    
def overpayment_messages (row):
    
    """рассчитывает доход, сверх тарифного плана (сообщения)"""
    
    messages_overrun = row['messages_overrun']
    rub_per_message = row['rub_per_message']
    
    if messages_overrun > 0:

        
            return messages_overrun * rub_per_message

    
    else:
        
        return 0       
```

```python
complete_data['overpayment_mb'] = complete_data.apply(overpayment_mb, axis = 1).round(2)
complete_data['overpayment_calls'] = complete_data.apply(overpayment_calls, axis = 1).round(2)
complete_data['overpayment_messages'] = complete_data.apply(overpayment_messages, axis = 1).round(2)
```

```python
complete_data['pay_for_everything'] = (
    complete_data['overpayment_mb'] + complete_data['overpayment_calls']
    + complete_data['overpayment_messages'] + complete_data['rub_monthly_fee'])
```

## <span style="color: #228B22">6. Средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются</span><br>

Выдвинем гипотезу, что выручки тарифов "Ультра" и "Смарт" равны.

```python
alpha = 0.05
```

Проверим выдвинутую гипотезу

```python
results = st.ttest_ind(
    ultra_data['pay_for_everything'], smart_data['pay_for_everything']) 

print('p-значение:', results.pvalue)

if (  results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Проверка данной гипотезы подтверждает замечанное нами на графиках явление - пользователи "Смарт", вынужденые выходить за рамки своего строгого тарифа, все же в своем большинстве не сравниваются по размерам платежа с пользователями тарифа "ультра".

## <span style="color: #228B22">7. Выручка пользователей из Москвы отличается от выручки пользователей из других регионов</span><br>

Выдвинем гипотезу, что пользователи из Москвы по своему выходу за рамки тарифного плана отличаются от пользователй из других регионов. для этого, в первую очередь, выделим две группы и, в последствии, сравним их.


```python
moscow_mobile = clean_mobile_data.query('city == "Москва"')
another_city = clean_mobile_data.query('city != "Москва"')
```

```python
results = st.ttest_ind(
    moscow_mobile['pay_for_everything'], another_city['pay_for_everything']) 

print('p-значение:', results.pvalue)

if (  results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```
